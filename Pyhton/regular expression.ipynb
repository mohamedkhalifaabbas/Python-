```python
# matches contains objects each object conatian :  
# span(start index , end index) // tuple      match = "what i search"
# like : # <re.Match object; span=(0, 3), match='abc'> 
```

```python
def findmatch(pattern , text) :
    pattern = re.compile(pattern)
    matches = pattern.finditer(text) 
    for match in matches : 
        print(match)
```

```python
findmatch("abc", mytext)
```

```python
pattern = re.compile(r"abc") # search format
matches = pattern.finditer(mytext) 
results =[]
for match in matches : 
    start = match.span()[0]
    end = match.span()[1]
    results.append(mytext[start :end])
results    
```

```python
emails = '''
CoreyMSchafer@gmail.com
corey.schafer@university.edu
corey-321-schafer@my-work.net
'''

pattern = re.compile('[a-zA-Z0-9-.]+@[a-zA-Z-]+\.(com|edu|net)')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
emails = '''
CoreyMSchafer@gmaail.co)m
corey.schafer@university.ed)u
corey-321-schafer@my-work.ne)t
'''

pattern = re.compile(r'[\w.+-]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]{2,}') # {2,} 2 or more
matches = pattern.finditer(emails)
for match in matches:
    print(match)
length  = sum(1 for  i in matches)
print(length)
```
